# Detecting and treating outliers(IQR)

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - .5 * IQR
    upper_bound = Q3 + .5 * IQR

    # Return a filtered DataFrame that excludes outliers
    return data[(data[column] >= lower_bound) &  (data[column] <= upper_bound)]

# Start with the original data
cleaned_data = data.copy()  # Create a copy for cleaning

# Remove outliers for each numeric column
for column in numeric_columns:
    cleaned_data = remove_outliers_iqr(cleaned_data, column)

# Visualize the cleaned data again to confirm outliers are removed
for column in numeric_columns:
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(cleaned_data[column], kde=True)
    plt.title(f'Histogram of {column} after outlier removal')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=cleaned_data[column])
    plt.title(f'Boxplot of {column} after outlier removal')

    plt.show()
```

##### When to Use SMOTE

* The target variable is imbalanced.
* You want to improve the model's ability to predict the minority class.

### 1. Support Vector Machine (SVM)
#### Explanation:
* SVM is a supervised learning algorithm that finds the optimal hyperplane to separate data points into different classes. It is especially effective for datasets where the classes are well-separated in high-dimensional space.
* It can handle both linear and non-linear relationships using kernels (e.g., polynomial or RBF kernels).
#### Why for Mushroom Dataset:
* SVM works well when the data is linearly separable or when class separation requires complex decision boundaries.
* It can effectively classify mushrooms as edible or poisonous based on distinct categorical features like cap shape, gill spacing, and others.
### 2. Decision Tree
#### Explanation:
* A Decision Tree splits the data into branches based on feature values, creating a tree-like structure. It uses conditions to predict the class of a sample.
* It’s simple to understand, interpretable, and can handle both categorical and numerical features.
#### Why for Mushroom Dataset:
* Mushroom data contains both categorical and continuous features, which Decision Trees handle well.
* Decision Trees are interpretable, allowing us to understand why certain features (e.g., gill color or stem surface) lead to specific classifications.
### 3. Naive Bayes
#### Explanation:
* Naive Bayes is a probabilistic classifier based on Bayes’ theorem, assuming features are conditionally independent given the class.
* It is fast, requires less training data, and is effective for categorical data.
#### Why for Mushroom Dataset:
* Since the dataset has many categorical features (e.g., cap shape, ring type), Naive Bayes can efficiently classify mushrooms.
* It’s useful when the features are assumed to contribute independently to the class label.
### 4. Random Forest
#### Explanation:
* Random Forest is an ensemble learning method that builds multiple Decision Trees and aggregates their outputs (e.g., majority voting for classification).
* It is robust to overfitting and performs well even with noisy or imbalanced data.
#### Why for Mushroom Dataset:
* Random Forest is well-suited for feature-rich datasets like mushrooms, where certain features might dominate classification (e.g., spore print color or stem height).
* It provides feature importance, helping identify which features are most critical for classification.
### 5. K-Nearest Neighbors (KNN)
#### Explanation:
* KNN is a simple, non-parametric algorithm that classifies samples based on the class of their nearest neighbors in the feature space.
* It relies on distance metrics (e.g., Euclidean or Manhattan) to identify neighbors.
#### Why for Mushroom Dataset:
* KNN works well for datasets where similar samples (e.g., mushrooms with similar cap shape or color) belong to the same class.
* It’s intuitive and can adapt to patterns in the data, but it may require feature scaling for continuous variables like cap diameter.
### 6. Gradient Boosting
#### Explanation:
* Gradient Boosting is an ensemble method that builds models sequentially, where each model corrects the errors of the previous one. Popular implementations include XGBoost, LightGBM, and CatBoost.
* It’s highly flexible and achieves state-of-the-art performance in many classification tasks.
#### Why for Mushroom Dataset:
* Gradient Boosting is ideal for complex datasets like mushrooms, as it captures non-linear relationships between features.
* It handles categorical and numerical features well, making it suitable for the diverse nature of the mushroom dataset.


```python
# Define models
models = {
    'Support Vector Classifier' : SVC(),
    'Decision Tree Classifier' : DecisionTreeClassifier(),
    'Random Forest Classifier' : RandomForestClassifier(),
    'Gradient Boosting Classifier' : GradientBoostingClassifier(),
    'Gaussian Naive Bayes' : GaussianNB(),
    'K-Nearest Neighbors' : KNeighborsClassifier()
}


# Store accuracies
train_accuracies = []
test_accuracies = []

# Training models
for model_name, model in models.items():

  # fit the model on training data
  model.fit(x_train_scaled,y_train)
  train_accuracy = model.score(x_train_scaled, y_train)
  train_accuracies.append(train_accuracy)
```

# 12. **Model evaluation:**

```python
for model_name, model in models.items():
  # predict on test data
  y_pred = model.predict(x_test_scaled)
  test_accuracy = model.score(x_test_scaled, y_test)
  test_accuracies.append(test_accuracy)

  print(f"Model: {model_name}")
  print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
  print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
  con = confusion_matrix(y_test, y_pred)
  plt.figure(figsize=(8, 6))
  sns.heatmap(con, annot=True, fmt=".2f", cmap='plasma', linewidths=0.5)
  plt.show()
```

# 13. **Hyperparameter Tuning:**

```python
best_accuracy = 0
best_model = 0
# Define hyperparameter grids for each model
param_grids = {
    
    'Support Vector Classifier': {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    },
    'Decision Tree Classifier': {
        'max_depth': [None, 5, 10, 20],
        'min_samples_split': [2, 5, 10]
    },
    'Random Forest Classifier': {
        'n_estimators': [50, 100],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5]
    },
    'Gradient Boosting Classifier': {
        'n_estimators': [50, 100],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5]
    },
    'Gaussian Naive Bayes': {},
        'K-Nearest Neighbors': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance']
    }
}

# Train and tune each model using Grid Search
for model_name in param_grids.keys():
    if model_name == 'Gaussian Naive Bayes':
        model = GaussianNB()
    elif model_name == 'Support Vector Classifier':
        model = SVC()
    elif model_name == 'Decision Tree Classifier':
        model = DecisionTreeClassifier()
    elif model_name == 'Random Forest Classifier':
        model = RandomForestClassifier()
    elif model_name == 'Gradient Boosting Classifier':
        model = GradientBoostingClassifier()
    elif model_name == 'K-Nearest Neighbors':
        model = KNeighborsClassifier()
    else:
        # Initialize the appropriate classifier based on the name
        model_class = eval(model_name.replace(" ", ""))
        model = model_class()

    # Set up GridSearchCV with cross-validation
    grid_search = GridSearchCV(model, param_grids[model_name], cv=5)

    # Fit the model on training data
    grid_search.fit(x_train_scaled, y_train)

    # Predict on test data using the best estimator
    y_pred = grid_search.best_estimator_.predict(x_test_scaled)

    # Calculate accuracy on test data
    test_accuracy = grid_search.score(x_test_scaled, y_test)

    # Print results
    print(f"Model: {model_name}")
    print(f"Best Parameters: {grid_search.best_params_}")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}")

    # Check if this is the best model so far
    if test_accuracy > best_accuracy:
        best_accuracy = test_accuracy
        best_model = grid_search.best_estimator_
```
